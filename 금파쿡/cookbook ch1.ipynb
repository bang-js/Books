## FinanceDataReader
FRED data 가져오기\
DataReader의 상위버전

```python
import FinanceDataReader as fdr
```

```python
import matplotlib.pyplot as plt
```

미국 자료

```python
df = fdr.DataReader('FRED:ICSA', data_source='fred', start='1950-1-1') # start 안잡으면 1980년부터
df
```

```python
df = fdr.DataReader('FRED:NASDAQCOM,DFF', data_source='fred', start='1950-1-1')
df
```

```python
df['LOG_NASDAQCOM'] = np.log(df['NASDAQCOM'])
df['PC_NASDAQCOM'] = df['NASDAQCOM'].pct_change()
```

```python
fig, axes = plt.subplots(figsize=(50,10))

line1 = axes.plot(df.index, df['LOG_NASDAQCOM'], color='green', label='LOG_NASDAQCOM') #, linestyle='dashed'
axes1 = axes.twinx()
line2 = axes1.plot(df.index, df['DFF'], label='DFF', color='orange') #, 

lines = line1 + line2
labels = [l.get_label() for l in lines]
axes.legend(lines, labels)
plt.show()
```
