# 1.1 인공지능과 머신러닝


## 1.1 인공지능과 머신러닝, 딥러닝

### 머신러닝과 딥러닝

- **머신러닝**은 프로그램 되는 것이 아니라 **훈련**되는 것
- 많은 샘플 속에서 통계적 구조를 찾아 그 작업을 **자동화하기 위한 규칙**을 만들어 냄 (ex. 여행 사진 tagging을 자동화하고 싶다면 이미 사람들이 tag해놓은 다수의 사진 샘플을 ML시스템에 제공해서 특정 사진에 태그를 연관시키는 규칙을 학습)

image.png

- 대량의 복잡한 데이터셋을 다루기에 Bayesian analysis 같은 전통적인 통계분석을 적용하기 힘듬, 따라서 이론보다는 실습지향적

머신러닝 알고리즘은 샘플과 기대값이 주어졌을 때 데이터 처리 작업을 위한 실행 규칙을 찾는 것, 따라서 이하 3가지가 필요
1.  input data point : 입력하는 데이터
2. expactational output : 각 입력값에 대응되는(요구되는) 출력값
3. measure : 알고리즘 성능을 측정하는 방법, 알고리즘의 현재 출력과 기대 출력 차이를 측정/결정하기 위해 필요하며 작동방식에 피드백을 주어 이를 교정(학습)

- ML은 의미 있는 데이터로의 **변환**이자, 입력 데이터를 기반으로 기대 출력에 가깝게 만드는 유용한 **표현(representation)**을 학습하는 것 → 표현이란 데이터를 인코디할 때 데이터를 바라보는 다양한 방법을 의미\
ex. RGB vs HSV → 목표가 무엇인지에 따라 사용해야 하는 표현이 달라짐
- 머신러닝은 가능성 있는 공간(가설 공간, hypothesis space)을 사전에 정의하고 피드백 신호의 도움을 받아 입력 데이터에 대한 유용한 변환을 찾는 것


- **딥러닝**은 ML은 한 분야로 연속된 **층(layer)**에서 점진적으로 의미 있는 표현을 배우는 데 강점을 가진, 데이터로부터 표현을 학습하는 새로운 방식
- 딥러닝의 딥은 그저 연속된 층으로부터 표현을 학습한다는 개념

image.png

최종 출력에 대해 점점 더 많은 정보를 가지지만 원본 이미지와는 점점 더 다른 표현으로 숫자 이미지가 변환됨\
즉 딥러닝은 데이터 표현을 학습하기 위한 다단계 처리 방식

### 딥러닝의 작동원리

- 어떤 층에서 일어나는 변환은 그 층의 가중치를 파라미터로 가지는 함수로 표현
- 따라서 'LEARNING'은 주어진 input을 정확한 target에 mapping하기 위해 신경망의 모든 층에 있는 weights를 찾는 작업을 의미 
- 다만 수백, 수만, 수백만의 파라미터를 가지기에 모든 파라미터의 정확한 값을 analytic하게 찾을 수 없으며 파라미터 하나를 바꾸면 나머지 모든 파라미터에 영향을 미침



image.png

- 신경망의 output을 제어하려면 출력이 예상보다 얼마나 얼마나 벗어났는지를 측정해야함
- 이를 위해 loss ftn 또는 objective ftn을 계산(한 샘플에 대한 예측과 실제값의 차이를 점수로 계산)
- 딥러닝은 이 점수를 피드백 신호로 받아들여 현재 샘플의 손실 점수가 감소되는 방향으로 가중치 값을 조금씩 수정하는 것
- 이러한 작업은 Backpropagation 알고리즘을 구현한 Optimizer가 담당 : 초기에는 가중치가 랜덤으로 할당, 이후 네트워크가 데이터를 반복적으로 처리하면서 가중치가 손실 점수를 최소화하는 방향으로 점차 조정됨

ML/DL이 할 수 있는 것들 : 이미지 분류, 필기/음성 인식, 기계 번역, Text-To-Speech, 디지털 비서, 광고 타게팅, 자율주행, 질문에 대답하는 능력, 바둑, 주식 투자

# 1.2 딥러닝 이전

확률적 모델링
- Naive Bayes : 입력 데이터의 특성이 모두 독립이라 가정, 베이즈 정리를 적용하는 ML, 1950년대부터 사용
- Logistic Regression : **분류** 알고리즘

kernel
- 대표적으로 SVM(서포트 벡터 머신) : 2개의 다른 범주에 속한 데이터 포인트 그룹을 잘 나눌 수 있는 좋은 결정 경계(decision boundary)를 찾는 기법
- 결정 경계가 하나의 hyperplane으로 표현될 수 있는 새로운 고차원 표현으로 데이터를 매핑 → 초평면돠 각 클래스의 가장 가까운 데이터 포인트 사이의 거리가 최대가 되는 최적의 결정 경계를 찾음
- SVM은 강력했지만 대용량 데이터셋에 확장되기 어려움, 이미지 분류 등에 적용하려면 수동으로 유용한 표현을 추출하는 특성 공학 단계를 거쳐야 하는데 매우 어렵고 불안정한 문제

Decision Tree
- 플로차트 같은 구조를 가지며 입력에 대해 출력을 예측
- 이외에도 Random Forest, Gradient Boosting이 있음


딥러닝의 특징
- ML 기법들은 간단한 변환을 통해 하나 또는 2개의 연속된 표현 공간으로만 변환(**얕은 학습방법**), 하지만 복잡한 문제에 필요한 잘 정제된 표현은 일반적으로 이런 방식으로 얻을 수 없음
- ML은 초기 입력 데이터를 여러 방식으로 변환하여 수동으로 좋은 표현을 찾는 특성 공학 단계를 거쳐야 하는데 DL은 이 단계를 완전 자동화함, 즉 한 번에 모든 특성을 학습할 수 있음
- 만약 얕은 학습방법을 연속적으로 사용하면 각 층의 효과는 빠르게 줄어듦, 이는 각 층들이 독립적으로 학습하기 때문에 발생, 2개 층을 가진 모델에서 최적의 첫 번째 표현 층은 하나의 층을 가진 모델에서의 최적의 첫 번째 표현 층과 달라야함
- DL은 표현 층을 순차적(greedily)이 아니라 동시에 공동으로 학습하게 만듦, 즉 모델의 모든 변화가 최종 목표를 향해 이루어짐

DL의 중요한 특징
1. 층을 거치면서 점진적으로 보다 복잡한 표현이 만들어진다는 것
2. 점진적 중간 표현이 공동으로 학습됨 : 각 층은 다른 층의 표현이 변함에 따라서 함께 바뀜


최근에는 구조적 데이터는 XGBoost(gradient boosting), 이미지 분류 같은 지각 문제는 DL을 사용

# 1.3 왜 딥러닝일까? 왜 지금일까?

- 합성곱 신경망과 역전파는 이미 1989년에 소개됨, LSTM 또한 1997년에 개발, 그런데 그동안 잠잠하다가 2010년대에 들어서 급부상
- '하드웨어(GPU), 데이터셋과 벤치마크, 알고리즘 향상'이라는 세 팩터가 발달하면서 ML이 본격적으로 발전



하드웨어
- NVIDIA와 AMD는 CPU를 대체할 수 있는 소형의 대용량 고속 병렬 칩(**GPU**)을 연구개발, 2007년 NVIDIA가 자사의 GPU 제품을 위한 프로그래밍 인터페이스인 CUDA(https://developer.nvidia.com/about-cuda)를 출시, 대기업은 NVIDIA Tesla K80처럼 딥러닝을 위해 개발된 GPU 수백 개로 구성된 클러스터에서 딥러닝 모델을 훈련
- Google은 **TPU**(텐서처리장치)를 2016년에 발표, 고성능 GPU보다 10배 이상 빠름

데이터
- Flicker에서 사용자가 붙인 이미지 태그는 컴퓨터 비전에서 중요한 데이터
- Youtube 또한 대용량 데이터셋
- Wikipedia는 자연어 처리에 중요
- ImageNet, Kaggle

알고리즘
- SVM, Random Forest 등의 얕은 학습방법은 깊게 쌓은 층을 통과해서 gradient를 전파하는 것이 가장 큰 문제, 신경망을 훈련하기 위한 피드백 신호가 층이 늘어남에 따라 약해지기 때문
- 신경망 층에 알맞는 activation ftn, 사전훈련을 불필요하게 하는 가중치 초기화 방법, RMSprop/Adam과 같은 좋은 최적화기법 등 중요 알고리즘이 개선되면서 gradient를 잘 전파되게 만들고 10개 이상의 층을 가진 모델을 훈련시킬 수 있게 되자 DL이 빛을 발함
- 이후 batch normalization, residual connection, depthwise separable convolution 등 고급기술이 개발되면서 층의 깊이 수천개인 모델을 훈련시킬 수 있게 됨

ML/DL에서 end-to-end learning이란 단어를 종종 발견하게 된다. end-to-end는 '끝에서 끝을 잇는'이라는 의미인데 end-to-end learning은 입력에서 출력까지 파이프라인 네트워크 없이 신경망으로 한 번에 처리한다는 의미이다. 파이프라인 네트워크란 전체 네트워크를 이루는 부분적인 네트워크이다\
  예를 들어, 기존의 Speech recognition system은 MFCC로 음성 파일의 특징 추출 → ML 알고리즘으로 음소를 알아냄 → 음소들로 단어를 만듦 → words 출력 같은 복잡한 과정을 거치지만 end-to-end learning은 음성 파일에서 바로 출력을 구할 수 있다.\
  참고로 ML/DL 논문에서 end-to-end trainable neural network이란 모델의 모든 매개변수가 하나의 손실함수에 대해 동시에 훈련되는 경로가 가능한 네트워크를 뜻한다. 즉, 신경망의 입력 및 출력을 직접 고려하여 네트워크 가중치를 최적화 될 수 있다.


C++과 CUDA를 할 줄 몰라도 파이썬과 자동미분을 지원하는 심볼릭 텐서 조작 프레임워크인 Theano와 Tensorflow를 통해 간단하게 모델을 구현할 수 있음
